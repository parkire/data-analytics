#### 1) Count of crises that African countries have experienced.
A count of 1 = Economic crisis
A count of 0 = No economic crisis

```python
crisis = df.groupby(['country', 'systemic_crisis']).size().unstack().fillna(0)
# crisis['Total'] = crisis.sum(axis = 1)
crisis.sort_values(by = 1, ascending = False)
```

#### A bar plot of the crisis

```python
crisis.sort_values(by = 1, ascending = False).plot(kind = 'bar')
```

Central African Republic (CAR) has had the highest number of system crises(19). South Africa has had no economic crisis.


### A trend of Kenya's crisis

```python
ke = df[(df.country == 'Kenya') & (df.systemic_crisis == 1) & (df.year)]
trend = ke.groupby(['systemic_crisis', 'year']).size()
trend
```

Kenya experienced a systemic crisis in the year: 1985, 1987, 1989, 1993, 1995, 1995 and 1999.

ii) Trends of Sovereign External debt default
Default = 1
No default  = 0

```python
crises = df.groupby(['country', 'sovereign_external_debt_default']).size().unstack().fillna(0)
crises.sort_values(by = 1, ascending = False).plot(kind = 'bar')
```

Kenya's Sovereign External debt default

```python
ke_default = df[(df.country == 'Kenya') & (df.sovereign_external_debt_default == 1) & (df.year)]
default = ke_default.groupby(['sovereign_external_debt_default', 'year']).size()
default
```

Kenya defaulted on extaernal loan payment in 1995 and 1997

#### Bank Crises in relation to Economic crisis

#### Bank crisis

```python
banks = df.groupby(['country','banking_crisis']).size().unstack()
banks.sort_values(by = 'crisis', ascending = False).plot(kind = 'bar')
```

### Bank crisis in Relation to Economic Crisis

```python
bank = df.groupby(['country','systemic_crisis', 'banking_crisis']).size().unstack().fillna(0)
bank.sort_values(by = 'crisis', ascending  = False)
```
