Analysis of Kenya's Olympics perfomance up to 2012

```python
import pandas as pd
df = pd.read_csv('summer.csv')
df
```

```python
kenya = df[df.Country == 'KEN' ]
```

Kenya's Olympics

```python
kenya
```

### History of Medals

```python
medals = df.groupby(['Medal']).size()
medals
```

Kenya had a total number of 10486 gold, 10310 silver and 10369 medals.

## The Kenyan athlete with the most medals in History of Olympics

```python
athlete = kenya.groupby(['Athlete', 'Medal']).size().unstack()
athlete = athlete.fillna(0)
athlete['Total'] = athlete.sum(axis=1)
athlete.sort_values(by = 'Total', ascending  = False)
# athlete['Total'] = df.sum('Bronze', 'Gold', 'Silver')
```

### Kenyan Perfomance in Olympics history

```python
perfomance = kenya.groupby(['Year']).size().sort_values(ascending = False)
perfomance
```

```python
Kenyas 2008 performance was the best in the countrys  Olympics participation.
```

### A plot of Kenyan medals in the Olympics

```python
history = kenya.groupby(['Year']).size().plot(y = 'Medal')
history
```

### A rank of Kenya in the Olympics

```python
rank = df.groupby(['Country']).size().sort_values(ascending = False)
rank = rank.reset_index(name='Count')
top_fifty = rank.head(50)
top_fifty
```
